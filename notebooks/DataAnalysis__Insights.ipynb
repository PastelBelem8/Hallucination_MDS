```python
import pandas as pd

from collections import defaultdict
from readers import read_json
```

## News domain


```python
all_insights_news = defaultdict(list)

for i in range(1, 6):
    dnews = read_json(f"../data/SummHay_news/raw/topic_news{i}.json")
    subtopics_news = {subtopic["subtopic_id"]: subtopic for subtopic in dnews["subtopics"]}
    insights_news = {}
    for subtopic in subtopics_news.values():
        for insight in subtopic["insights"]:
            insights_news[insight["insight_id"]] = insight
            all_insights_news["topic"].append(dnews["topic"])
            all_insights_news["topic_metadata"].append(dnews["topic_metadata"])
            all_insights_news["subtopic_name"].append(subtopic["subtopic_name"])
            all_insights_news["subtopic"].append(subtopic["subtopic"])
            all_insights_news["subtopic_uuid"].append(subtopic["subtopic_id"])
            all_insights_news["insight_name"].append(insight["insight_name"])
            all_insights_news["insight"].append(insight["insight"])
            all_insights_news["insight_uuid"].append(insight["insight_id"])

all_insights_news = pd.DataFrame(all_insights_news)
all_insights_news.to_csv("../data/SummHay_news/insights.csv", index=None)
```

## Conversation domain

```python
all_insights_conv = defaultdict(list)

for i in range(1, 6):
    dconv = read_json(f"../data/SummHay_conv/raw/topic_conv{i}.json")
    subtopics_conv = {subtopic["subtopic_id"]: subtopic for subtopic in dconv["subtopics"]}
    insights_conv = {}
    for subtopic in subtopics_conv.values():
        for insight in subtopic["insights"]:
            insights_conv[insight["insight_id"]] = insight
            all_insights_conv["topic"].append(dconv["topic"])
            all_insights_conv["topic_metadata"].append(dconv["topic_metadata"])
            all_insights_conv["subtopic_name"].append(subtopic["subtopic_name"])
            all_insights_conv["subtopic"].append(subtopic["subtopic"])
            all_insights_conv["subtopic_id"].append(subtopic["subtopic_id"])

            all_insights_conv["insight_name"].append(insight["insight_name"])
            all_insights_conv["insight"].append(insight["insight"])
            all_insights_conv["insight_id"].append(insight["insight_id"])

all_insights_conv = pd.DataFrame(all_insights_conv)
all_insights_conv.to_csv("../data/SummHay_conv/insights.csv", index=None)
all_insights_conv
```

```python
dconv = read_json(f"../data/SummHay_conv/raw/topic_conv1.json")
subtopics_conv = {subtopic["subtopic_id"]: subtopic for subtopic in dconv["subtopics"]}
insights_conv = {}
for subtopic in dconv["subtopics"]:
    for insight in subtopic["insights"]:
        insight["subtopic_id"] = subtopic["subtopic_id"]
        insight["subtopic"] = subtopic["subtopic"]
        insights_conv[insight["insight_id"]] = insight


overlap_matrix = defaultdict(list)
for d in dconv["documents"]:
    overlap_matrix["document_uuid"].append(d["document_id"])
    for c in d.keys():
        if c.startswith("label_"):
            overlap_matrix[c.replace("label_", "")].append(d[c])

overlap_matrix = pd.DataFrame(overlap_matrix).set_index("document_uuid")
subtopic_cols = list(subtopics_conv.keys())
insight_cols = list(insights_conv.keys())

# How many documents have overlapping subtopics
(overlap_matrix[subtopic_cols] != 0).sum(axis=0)
```

```python
# Note however, that documents may have overlapping subtopics but be
# of different insights. So in practice, we'd like to know how many
# insights are overlapping
overlap_matrix[insight_cols].sum(axis=0).value_counts()
```

```python
# -------------------------------------------
# let us estimate the number of combinations
# we could find using this dataset
# -------------------------------------------
from itertools import combinations

def get_subtopic(ins):
    return insights_conv[ins]["subtopic_id"]

unique_combinations = defaultdict(lambda: 0)
for ins_id in insight_cols:
    subtopic_uuid = get_subtopic(ins_id)
    doc_uuids = overlap_matrix[overlap_matrix[ins_id] != 0][ins_id].index

    if len(doc_uuids) == 0:
        continue
        
    for comb in combinations(doc_uuids, 2):
        comb = [subtopic_uuid] + sorted(comb) 
        unique_combinations[tuple(comb)] += 1

# These are the total number of combinations w/o 
# the restriction that there must be at least 2
# shared subtopics
print("Unique Combs:", len(unique_combinations))
print("Combs w/ 2+ shared insights:", len([combs for combs, counts in unique_combinations.items() if counts > 1]))
```
